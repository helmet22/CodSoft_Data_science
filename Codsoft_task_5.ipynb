```python
# importing required libraries 
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 

# for warning error filtering 
import warnings 
warnings.filterwarnings('ignore')
```

```python
# display dataset 
data = pd.read_csv('creditcard.csv')

data.head(10)
```

This dataset presents transactions that occurred in two days, where we have 492 frauds out of 284,807 transactions. The dataset is highly unbalanced, the positive class (frauds) account for 0.172% of all transactions.

The feature 'Amount' is the transaction Amount, this feature can be used for example-dependant cost-sensitive learning. Feature 'Class' is the response variable and it takes value 1 in case of fraud and 0 otherwise.

```python
data.info()
```

```python
print("Dataset has "+str(len(data))+" observation ")
print ("Dataset  has "+str(len(data.columns))+" variables or payments ")
```

```python
# checking any null values 
data.isnull().sum()
```

The dataset does not  have any null values.
So, we will check for unique  values of traget output 

```python
data['Class'].unique()
data['Class'].value_counts()
```

As our dataset is unbalanced . So , we will create new balanced  dataset including all fraud transaction and  normal transaction 

```python
normal_transaction = data[data["Class"]==0]
fraud_transaction  = data[data["Class"]==1]
normal_dataset = normal_transaction.sample(n=500,random_state=50, replace = False )
fraud_dataset = fraud_transaction.sample( n=492, replace = False)

# display new normal transaction  dataset
normal_dataset.head()
```

```python
# display fraud transaction dataset
fraud_dataset
```

After creating the dataset of almost equal number of observation , we will  merge it and form one new dataset 

```python
new_data = pd.concat([normal_dataset,fraud_dataset],axis=0)
# display new dataset 
new_data.head()
```

```python
new_data.describe().T
```

 We head to model selection and training 

```python
# import required model from sklearn 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score
```

```python
x = new_data.drop(['Class'],axis=1)
y = new_data[['Class']]

```

```python
X_train,X_test,Y_train,Y_test = train_test_split(x,y,test_size = 0.2,random_state=50)
```

```python
# Logistic Regression model 
model_1 = LogisticRegression()
model_1.fit(X_train,Y_train)
Y_predict_1 = model_1.predict(X_test)
# Evaluating model performance 
accuracy_score_1 = accuracy_score(Y_test,Y_predict_1)
precision_score_1 = precision_score(Y_test,Y_predict_1)
recall_score_1 = recall_score(Y_test,Y_predict_1)
f1_score_1 = f1_score(Y_test,Y_predict_1)

# display the model performance 
print(" Logistic Regression Accuracy: ",accuracy_score_1)
print(" Logistic Regression Precision: ",precision_score_1)
print(" Logistic Regression recall: ", recall_score_1)
print(" Logistic Regression F1-score: ",f1_score_1)

```

```python
# Random forest model 
model_2 = RandomForestClassifier(n_estimators=100,random_state= 42)
model_2.fit(X_train,Y_train)
Y_pred_2 = model_2.predict(X_test)

# Evaluating  classifier model performance 
accuracy_score_2 = accuracy_score(Y_test,Y_pred_2)
precision_score_2 = precision_score(Y_test,Y_pred_2)
recall_score_2 = recall_score(Y_test,Y_pred_2)
f1_score_2 = f1_score(Y_test,Y_pred_2)

# display classifier model performance 
print(" Logistic Regression Accuracy: ",accuracy_score_2)
print(" Logistic Regression Precision: ",precision_score_2)
print(" Logistic Regression recall: ", recall_score_2)
print(" Logistic Regression F1-score: ",f1_score_2)

```
